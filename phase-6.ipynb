## Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": [
        "2026-01-05", "2026-01-12", "2026-01-18",
        "2026-02-03", "2026-02-10", "2026-02-17",
        "2026-03-02", "2026-03-09", "2026-03-16",
        "2026-04-04", "2026-04-11", "2026-04-18",
        "2026-05-06", "2026-05-13", "2026-05-20"
    ],
    "Product": [
        "Laptop", "Mobile", "Tablet",
        "Laptop", "Mobile", "Tablet",
        "Laptop", "Mobile", "Tablet",
        "Laptop", "Mobile", "Tablet",
        "Laptop", "Mobile", "Tablet"
    ],
    "Region": [
        "East", "West", "North",
        "South", "East", "West",
        "North", "South", "East",
        "West", "North", "South",
        "East", "West", "North"
    ],
    "Sales": [
        50000, 30000, 20000,
        55000, 32000, 22000,
        60000, 35000, 25000,
        58000, 37000, 24000,
        62000, 39000, 26000
    ],
    "Units": [
        5, 10, 8,
        6, 11, 9,
        7, 12, 10,
        6, 13, 9,
        8, 14, 11
    ],
    "Rep": [
        "John", "Alice", "Bob",
        "John", "Alice", "Bob",
        "John", "Alice", "Bob",
        "John", "Alice", "Bob",
        "John", "Alice", "Bob"
    ]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df, values='Sales', index='Region', columns='Product')
```

```python
pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='median')
```

```python
# Multiple vales
pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
```

```python
# Cross Tabs(Count)
pd.crosstab(df['Region'], df['Product'])
```
