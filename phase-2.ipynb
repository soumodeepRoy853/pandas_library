## DataFrames in Pandas

```python
import numpy as np
import pandas as pd
```

```python
# Creating dataFrames using dictionary with keys
data_dic = {
    "Name": ["Rit", "Soumo", "Tukai"],
    "Age": [25, 26, 24],
    "City": ["Mekhliganj", "Jalpaiguri", "Siliguri"],
    "Salary": [40000, 50000, 30000]
}

df = pd.DataFrame(data_dic)

# Creating dataFrames using list without keys
data_list = [
    ["Rit", 25, "Mekhliganj", 40000],
    ["Soumo", 26, "Jalpaiguri", 50000],
    ["Tukai", 24, "Siliguri", 30000],
]

df2 = pd.DataFrame(data_list)

# Insert keys into existing dataframes
keys = ["Name", "Age", "City", "Salary"]

df3 = pd.DataFrame(data_list, columns=keys)
```

## Selection and Indexing of columns

```python
# Selection of columns
df["Name"] #select one column
```

```python
# select multiple columns
df[["Name", "Salary"]]
```

```python
# Creating a new column into existing Dataframes
df["Occupation"] = ["Engg.", "Doc.", "Lawyer"]
df
```

## Removing a Column

```python
# Drop a columns
df.drop('Occupation', axis=1) # Its only show copy of actual data, do not drop the column from RAM. If want to drop column from RAM, use 'inplace=true'
```

```python
df.drop('Occupation', axis=1, inplace=True) # Drop column permanently
df
```

```python
# Drop multiple Columns
df.drop(["City", "Salary"], axis=1)
```

```python
# Drop Row
df.drop(0) #It always choose axis=0
```

## Selecting Rows

```python
# Select a row location
df.loc[0]
```

```python
# Select multiple row location
df.loc[[0, 1]]
```

```python
# Select a perticular rows of column
df.loc[[0, 1],["City", "Salary"]]
```

## Conditional Operation

```python
# Select people whose age is above 24
df[df["Age"] > 24] # Use only one condition
```

```python
# Use multiple condition
df[(df["Age"] > 24) & (df["Salary"] == 50000)]  
```
