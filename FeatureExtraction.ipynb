# Make a new column for First_Name and Last_name from Real Name


```python
df["Real Name"].head()
```

```python
#Extract First Name
df["First_Name"] = df["Real Name"].str.split().str[0]
df[["Real Name", "First_Name"]]
```

```python
#Extract Last Name
df["Last_Name"] = df["Real Name"].str.split().str[-1]
df[["Real Name", "First_Name", "Last_Name"]]
```

```python
# Function for character count
def char_leng(character):
    count = 0
    for i in character:
        count += 1
    return count
```

```python
df["Char_Length"] = df["Character"].apply(char_leng)
df[["Character","Char_Length"]]
```

```python
df.head(10)
```

```python
# Real Name word count
def count_words(name):
    return(len(name.split()))
```

```python
df["Word_Count"] = df["Real Name"].apply(count_words)
df[["Real Name", "Word_Count"]]
```

```python
# Encoding Role with 0 and 1
def encoded_role(role):
    if role == "Hero":
        return 2
    elif role == "Villain":
        return 0
    elif role == "Antihero":
        return 1
```

```python
df["Encode_Role"] = df["Role"].apply(encoded_role)
df[["Role", "Encode_Role"]]
```

```python
# Power level encoding
def encoding_power(power):
    if power == "Low":
        return 0
    elif power == "Medium":
        return 1
    elif power == "High":
        return 2
```

```python
df["Encoded_Power"] = df["Power Level"].apply(encoding_power)
df[["Power Level", "Encoded_Power"]]
```

```python
# Count Powers
def count_powers(power):
    return len(power.split(","))
```

```python
df["Count_Powers"] = df["Powers"].apply(count_powers)
df[["Powers", "Count_Powers"]]
```
