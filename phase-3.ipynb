## Finding Missing Data

```python
import numpy as np
import pandas as pd
```

```python
# Find NaN values from a list
data = {
    "A": [1, np.nan, 2, 3],
    "B": [2, 3, 5, np.nan],
    "C": [4, np.nan, 7, np.nan],
    "D": [5, 8, np.nan, np.nan]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
# Count NaN
df.isna().sum()
```

## Removing Missing Data

```python
df
```

```python
#Removing only NaN value rows  
df.dropna()
```

```python
df.dropna(thresh=3)
```

```python
df.dropna().any()
```

## Fill Missing values

```python
df
```

```python
df.fillna(0) # Fill 0 all the NaN values
```

```python
# Fill specific values to the column wise
val = {"A":10, "B": 20, "C": 30, "D": 40}
df.fillna(val)
```
