## Pandas Series
A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index. 

```python
import pandas as pd
import numpy as np
```

```python
# Creating Series
arr = np.array([10, 20, 40, 60, 80])
my_levels = ['S', 'O', 'U', 'M', 'O']

dic = {1:10, 2:20, 3:30, 4:40}
```

```python
pd.Series(arr)
```

```python
pd.Series(arr, my_levels)
```

```python
pd.Series(dic)
```
