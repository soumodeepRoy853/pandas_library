## Merging Joining Concatination

```python
import numpy as np
import pandas as pd
```

```python
# Merging Two data frames
employees = pd.DataFrame({ # employee information
    'emp_Id' : [1, 2, 3, 4, 5],
    'name': ['Anna', 'Peter', 'Bob', 'Jack', 'Oggy'],
    'dept': ["HR", 'IT', 'Finance', 'IT', 'HR']
})

salaries = pd.DataFrame({ # salary information
   'emp_Id' : [1, 2, 3, 6, 7],
   'salary': [60000, 70000, 65000, 80000, 65000],
   'bonus': [5000, 2000, 4000, 3000, 1500] 
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries) # Merge only common keys like employee_Id
```

```python
pd.merge(employees, salaries, on='emp_Id', how='inner')
```

```python
pd.merge(employees, salaries, on='emp_Id', how='outer')
```

```python
pd.merge(employees, salaries, on='emp_Id', how='left') # Do operation on employees basis
```

```python
pd.merge(employees, salaries, on='emp_Id', how='right') # Do operation on salaries basis
```

## Concatination of two Data Frames

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
# Concate df1, df2
pd.concat([df1, df2]) # column-wise
```

```python
pd.concat([df1, df2], axis=1) # row-wise
```

## Joining of Two Data Frames

```python
df1 = pd.DataFrame({
    'Name': ["Alice", "Bob", 'Jack']
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'Score': [88, 90, 89]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```
