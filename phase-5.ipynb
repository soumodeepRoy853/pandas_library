## GroupBy Aggeregation

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'C', 'B', 'A', 'B'],
    'Store': ['S1', 'S2', 'S3', 'S1', 'S3', 'S2'],
    'Sales': [100, 200, 150, 300, 250, 150],
    'Qty': [10, 12, 15, 10, 13, 10],
    'Date': pd.date_range('2026-06-01', periods=6)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
cat = df.groupby('Store')['Qty'].sum()
cat
```

```python
# GroupBy multiple columns
cat = df.groupby(['Category', 'Store'])['Sales'].sum()
cat
```

```python
cat = df.groupby(['Category', 'Store'])['Qty'].sum()
cat
```
